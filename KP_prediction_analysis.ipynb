```python
import pandas as pd
file_path = "KpNet_Package/results/kp_1h_results_predictions.csv"
# Load the CSV file into a DataFrame
predictions_df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to understand its structure
predictions_df.head()
```

```python
# date - timestamp
# Labels - Ground Truth Kp-Index
# Predictions - predicted Kp-Index
# Epistemic - epistemic uncertainty associated with each prediction
#    -Epistemic refers to uncertanity due to lack of knowledge or data. Its a measure of how confident the model is in its predictions
# Aleotoric - aleatoric uncertainty associated with each prediction.
#    -Aleatpric incertainity is inherent randomness or variabitlity in the data itself, which the model cannot reduce, even with infinte data.
```

```python
# Unit Key

# Kp index ranges from 1-9
# 1 indicates very little geomagnetic activity
# 9 indicated a sverve geomagnetic storm

#NaN in Epistemic and Aleatoric indecate that the model couldn't estimate these values properly.
```

```python

import matplotlib.pyplot as plt


# Extract relevant columns
date_column = pd.to_datetime(predictions_df["Date"])
labels = predictions_df["Labels"]
predictions = predictions_df["Predictions"]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(date_column, labels, label="Labels", marker="o")
plt.plot(date_column, predictions, label="Predictions", marker="x")
plt.xlabel("Date")
plt.ylabel("Kp Index")
plt.title("Kp Index Prediction for 1-hour Horizon")
plt.legend()
plt.grid(True)
plt.show()
```
